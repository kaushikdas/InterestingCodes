```python
def findLogestLine(filename):
    f = open(filename, 'r')
    longestline =''
    for line in f:
        if len(line) > len(longestline):
            longestline = line
        
    print('Longest line is:', longestline)
    f.close()
    
findLogestLine(r'D:\workspace\kaushik\github\InterestingCodes\Python\poem.txt')
```

```python
def printFileWithLineNumber2(filename):
    f = open(filename, 'r')
    i = 1
    for line in f:
        print(i, line, end='')
        i += 1
    f.close()
    
def printFileWithLineNumber(filename):
    f = open(filename, 'r')
    allLines = f.readlines()
    for i in range(1, len(allLines) + 1):
        print(i, allLines[i-1], end='')
    f.close()
    
printFileWithLineNumber(r'D:\workspace\kaushik\github\InterestingCodes\Python\poem.txt')

print('\n-------------')

printFileWithLineNumber2(r'D:\workspace\kaushik\github\InterestingCodes\Python\poem.txt')
```

```python
def remove_lowercase(file1,file2):
    f1=open(file1, 'r')
    f2=open(file2, 'w')
    for line in f1.readlines():
        if line[0].islower() == True:
            f2.write(line)
    f1.close()
    f2.close()

remove_lowercase(r'D:\workspace\kaushik\github\InterestingCodes\Python\poem1.txt',
                 r'D:\workspace\kaushik\github\InterestingCodes\Python\poem2.txt')

```

```python
import pickle

f = open(r'member.dat', 'wb')

choice = 'y'
while choice == 'y':
    member = {}  #dict()
    memberNo = int(input('Enter member no: '))
    memberName = input('Enter member name: ')
    member['MemberNo'] = memberNo
    member['MemberName'] = memberName
    pickle.dump(member, f)
    
    choice = input('Want to add more member? (y/n) ')

f.close()
```

```python
import pickle

f = open(r'member.dat', 'rb+')
try:
    while True:
        curpos = f.tell()
        member = pickle.load(f)
        print(member)
        member['MemberName'] = member['MemberName'].upper()
        f.seek(curpos)
        pickle.dump(member, f)
except EOFError:
    print('Read all records')
    
f.close()

f = open(r'member.dat', 'rb+')  
try:
    while True:
        member = pickle.load(f)
        print(member)
except EOFError:
    print('Read all records')
f.close()   
```
