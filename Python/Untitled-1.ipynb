````python
def CSVCOPY(sourcefile, targetfile):
    import csv
    sf = open(sourcefile, newline='')
    tf = open(targetfile, 'w', newline='')
    rdr = csv.reader(sf)
    rows = []
    for row in rdr:
        rows.append(row)
    wrtr = csv.writer(tf)
    wrtr.writerows(rows)
    sf.close()
    tf.close()
````

````python
import csv

print('WRITING A CSV AFTER OPENING IT W/O GIVING ANY VALUE FOR newline')
f1 = open('simple.csv', 'w')
w = csv.writer(f1)
L = [
    ['Name', 'Gender'],
    ['Kaushik', 'M'],
    ['Pratima', 'F'],
    ['Maharnab', 'M']
]
w.writerows(L)
f1.close()

f2 = open('simple.csv', 'rb')
lines1 = f2.read()
f2.close()
print('\n----\nOUTPUT of THE ABOVE CSV WHEN READ AS BINARY')
print(lines1, end='')
print('----')

f3 = open('simple.csv', 'r')
lines2 = f3.read()
f3.close()
print('\n----\nOUTPUT of THE ABOVE CSV WHEN READ AS TEXT')
print(lines2, end='')
print('----')

f4 = open('simple.csv', 'r')
lines3 = f4.readlines()
f4.close()
print('\n----\nOUTPUT of THE ABOVE CSV WHEN READ AS TEXT (LINES)')
print(lines3, end='')
print('----')
````

````python
# sorting list using list method sort()
l = list('apple')
l.sort()
print("List sorted in ascending order:", l)
l.sort(reverse=True)
print("List sorted in descending order:", l)
````

````python
# sorting list using built-in function sorted()
l = ['a', 'p', 'p', 'l', 'e']  # this is same as l = list('apple')
print("List sorted in ascending order:", sorted(l))
print("List sorted in descending order:", sorted(l, reverse=True))
````

````python
def PUSH_IN(L):
    global S
    S = []
    for i in L:
        if i % 3 == 0:
            S.append(i)
    print("Stack\n", S)

````
